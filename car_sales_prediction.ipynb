# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Getting the dataset

```python
url = "https://raw.githubusercontent.com/ThierrryScotto/car-sales-prediction/master/tabela-fipe-historico-precos.csv"
```

```python
dataset = pd.read_csv(url)
dataset
```

```python
dataset.describe()
```

# Data Preparation

```python
dataset.head(5)
```

```python
dataset.rename(columns={"codigoFipe"    : "FipeCode",
                        "marca"         : "Brands",
                        "modelo"        : "Model",
                        "anoModelo"     : "ModelYear",
                        "mesReferencia" : "ManufacturingMonth",
                        "anoReferencia" : "ManufacturingYear",
                        "valor"         : "Price" }, 
               inplace = True)
dataset.columns
```

```python
dataset.drop(columns=["Unnamed: 0"], inplace = True)
dataset
```
